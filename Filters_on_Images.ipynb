```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('filter.jpeg', cv2.IMREAD_GRAYSCALE)

# Apply Sobel filter (horizontal)
sobel_horizontal = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)

# Apply Sobel filter (vertical)
sobel_vertical = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

# Combined Sobel (Magnitude of gradients)
sobel_combined = cv2.magnitude(sobel_horizontal, sobel_vertical)

# Additional Filter - Gaussian Blur
gaussian_blur = cv2.GaussianBlur(img, (9, 9), 0)

# Display results
fig, axes = plt.subplots(1, 5, figsize=(20, 5))
axes[0].imshow(img, cmap='gray')
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(sobel_horizontal, cmap='gray')
axes[1].set_title('Sobel Horizontal')
axes[1].axis('off')

axes[2].imshow(sobel_vertical, cmap='gray')
axes[2].set_title('Sobel Vertical')
axes[2].axis('off')

axes[3].imshow(sobel_combined, cmap='gray')
axes[3].set_title('Sobel Combined')
axes[3].axis('off')

axes[4].imshow(gaussian_blur, cmap='gray')
axes[4].set_title('Gaussian Blur')
axes[4].axis('off')

plt.tight_layout()
plt.show()
```
